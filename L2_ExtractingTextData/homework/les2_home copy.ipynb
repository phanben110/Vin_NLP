## Câu hỏi 1: Lấy html của trang web
Hoàn thiện chương trình sau:
- Lấy html từ url sau: "http://www.hubertiming.com/results/2017GPTR10K"
- Tạo đối tượng BeautifulSoup từ html

```python
#### YOUR CODE HERE ####

url = 'https://www.hubertiming.com/results/2017GPTR10K'
# get the html from the url
html = urlopen(url)
# parse the html using beautiful soup and store in variable 'soup'
soup = BeautifulSoup(html, 'lxml')
#### END YOUR CODE ####
type(soup)
```

```python
#### YOUR CODE HERE ####

# get h2 tag
h2 = soup.find_all('h2')
# get title from soup
title = soup.title
print(title)
# get text from soup
text = soup.get_text()
# Get all images from soup
images = soup.find_all('img')
# Get all links from soup
links = soup.find_all('a')

#### END YOUR CODE ####
```

## Câu hỏi 3: Trích xuất các dữ liệu trong bảng và làm sạch

### Câu hỏi 3.1: Lấy ra hết các hàng trong bảng
Gợi ý: Đưa tham số 'tr' vào soup.find_all()

```python
#### YOUR CODE HERE ####

# get the rows from table
rows = soup.find_all('tr')
#### END YOUR CODE ####

# Print the first 10 rows for sanity check
print(rows[:10])
```

```python
# Print out nội dung của các hàng trong bảng
row_td = []
for row in rows:
    row_td.append(row.find_all('td'))
```

```python
str_cells = str(row_td[5:])
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

### Câu hỏi 3.2: Làm sạch dữ liệu trong bảng
Đến đây ta đã có dữ liệu của các hàng trong bảng, được lưu ở biến rows. Hãy viết hàm tiền xử lý các dữ liệu trong cells: loại bỏ các ký tự "\r", "\n", thay thế pattern "<.*?>" thành " ".

```python
import re
import ast

def clean_rows(rows_):
    #### YOUR CODE HERE ####
    # remove "\r", "\n" from rows
    rows_ = re.sub('\r', '', rows_)
    rows_ = re.sub('\n', '', rows_)
    # replace '<.*?>' with ' '
    rows_ = re.sub('<.*?>', ' ', rows_)
    return rows_   

    #### END YOUR CODE ####
list_rows = clean_rows(cleantext)
# convert list_rows to list of lists
list_rows = list_rows.split('], [')
# remove '[[' in the first, last element
list_rows[0] = list_rows[0].replace('[[', '')
list_rows[-1] = list_rows[-1].replace(']]', '')
# # convert list_rows to list of lists
list_rows = [row.split(',') for row in list_rows]
list_rows
```
