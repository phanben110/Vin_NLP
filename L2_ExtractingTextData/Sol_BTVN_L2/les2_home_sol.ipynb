Dữ liệu của bài toán này là một [web](https://www.hubertiming.com/results/2017GPTR10K) được trình bày dưới dạng bảng. Bài tập yêu cầu các kiến thức về lập trình Python với các thư viện BeautifulSoup, numpy, pandas, urlopen

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Tạo biến chứa url 
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
# Khởi tạo đối tượng BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the h2 tag
h2 = soup.h2
print(h2)
```

```python
# Print out the p tag
text = soup.p()
print(soup.text)
```

```python
# Truy cập tất cả các thẻ img
soup.find_all('img')
```

```python
# Print out các link liên kết
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# In ra 30 nội dung phần tử của thẻ td
rows = soup.find_all('td')
print(rows[:30])
```

```python
# Print out nội dung của các hàng trong bảng 
# Gợi ý: lấy ra thẻ tr trước sau đó trích xuất nội dung thẻ td
rows = soup.find_all('tr')
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
# print out nội dung text
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re


list_rows_ = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    str_cells = str_cells.replace("\r", "") # loại bỏ các ký tự "\r"
    str_cells = str_cells.replace("\n", "") # loại bỏ các ký tự "\n"
    # thay thế pattern "<.*?>" thành " "
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean,'',str_cells))
    list_rows_.append(clean2)   
```

```python
# Đưa list_rows vào pd.DataFrame()
df = pd.DataFrame(list_rows_)
df.head(10)
```

```python
# Tách cột "0" ở vị trí dấu ":" thành nhiều cột
df1 = df[0].str.split(',', expand=True)
df1.head(20)
```

```python
# Bỏ '[]' và ':' ở cột thứ 0 và xóa ] ở cột thứ 8
df1[0] = df1[0].str.strip('[]')
df1[0] = df1[0].str.strip(':')
df1[8] = df1[8].str.strip(']')
df1.head(20)
```

```python
# Truy cập button của bảng
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
# Trích xuất text giữa html tags cho button của bảng
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
# Chuyển danh sách trong headers vào một dataframe
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
# ta cần tách cột "0" thành nhiều cột xóa 2 dấu [ và ] ở cột 0 và 8
df3 = df2[0].str.split(',', expand=True)
df3[0] = df3[0].str.strip('[')
df3[8] = df3[8].str.strip(']')
df3.head()
```

```python
# Ghép 2 dataframe
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
# Gán hàng đầu tiên thành header
df5 = df4.rename(columns=df4.iloc[0])
df5.head(30)
```

```python
# Kiểm tra dữ liệu có missing values hay không
df5.info()
df5.shape
```

```python
# Bỏ đi các hàng dữ liệu có miss value
df6 = df5.dropna(axis=0, how='any')
```
