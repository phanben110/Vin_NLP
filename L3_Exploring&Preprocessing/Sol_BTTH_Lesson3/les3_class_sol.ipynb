## Câu hỏi 2: Xây dựng các hàm xử lý văn bản

### Câu hỏi 2.1: Xây dựng hàm làm sạch dữ liệu
Mô tả: hàm loại bỏ đi các ký số và xóa ký tự '-'

```python
# clean text
def clean_text(text):
#### YOUR CODE HERE ####
    text = re.sub(r'\d+', '', text)
    result = re.sub(r'\-','',text)
    return result
#### END YOUR CODE #####
```

### Câu hỏi 2.2: hàm chuyển văn bản về viết thường

```python
# make all text lowercase
#### YOUR CODE HERE ####
def text_lowercase(text):
    return text.lower()
#### END YOUR CODE #####
```

### Câu hỏi 2.3: Xây dựng hàm loại bỏ dấu chấm câu

```python
# remove punctuation
def remove_punctuation(text):
    #### YOUR CODE HERE #### 
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

    #### END YOUR CODE #####
```

 Câu hỏi 2.4: Tách từ

Gợi ý: Ở đây ta dùng dữ liệu tiếng việt nên sẽ dùng thư viện underthesea để tách

---



```python
# tokenize
def tokenize(text):
#### YOUR CODE HERE ####
    text = word_tokenize(text)
    return text
#### END YOUR CODE #####
```

### Câu hỏi 2.5: Loại bỏ từ dừng (stopwords)

```python
def remove_stopwords(strings):
    #### YOUR CODE HERE ####
    f = open('/content/drive/MyDrive/NLP/Đáp án cho bài tập thực hành/Lesson3/vietnamese-stopwords.txt', 'r')  
    stopwords = f.readlines()
    stop_words = [s.replace("\n", '') for s in stopwords]
    doc_words = []
    
    for word in strings:
        if word not in stop_words:
            doc_words.append(word)
    return doc_words
    #### YOUR CODE END HERE ####
```

### Câu hỏi 2.7: Xây dựng hàm tiền xử lý
Gợi ý: hàm này sẽ gọi các hàm đã được viết ở trên

```python

def preprocessing(text):
#### YOUR CODE HERE ####

    text = clean_text(text)
    text = text_lowercase(text)
    text = remove_punctuation(text)
    text = tokenize(text)
    text = remove_stopwords(text)
    text = ' '.join(text)
    return text

#### END YOUR CODE #####
```
